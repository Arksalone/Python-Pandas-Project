### Effort 3 - Visualize your data!
Import the Plotly Express library into the notebook

```python
# Import Plotly Express (use common alias)
import plotly.express as px
```

Create a bar chart of annual beer consumption by country

```python
# Create a bar chart of annual beer consumption by country
# Change the x label to "Country"
# Change the title to "Annual Beer Consumptions, Servings"
px.bar(x = countries, y = beer_servings, labels = {'x': 'Country', 'y': 'beer_servings'}, title = 'Annual Beer Consumptions, Servings')
```

Create a bar chart of annual total alcohol consumption by country

```python
# Create a bar chart of annual alcohol consumption by country
# Change the x label to "Country"
# Change the title to "Total Alcohol Consumption, Liters"
px.bar(x = countries, y = total_alcohol_liters, labels = {'x': 'Country', 'y': 'total_alcohol_liters'}, title = 'Total Alcohol Consumption, Liters')
```

Create a histogram of alcohol consumption in countries.

```python
# Histogram of alcohol consumption
# Change the x label to "Total Annual Alcohol Consumption, Liters"
# Change the title to "Histogram of Alcohol Consumption amoung Countries"
px.histogram(x = total_alcohol_liters, y = countries, labels = {'x': 'Total Annual Alcohol Consumption', 'y': 'countries'},
       title = 'Histogram of Alcohol Consumption amoung Countries')
```

Create a histogram of beer servings among countries.

```python
# Histogram of alcohol consumption
# Change the number of bins to 8
# Change the x label to "Total Annual Beer Consumption, Liters"
# Change the title to "Histogram of Beer Consumption amoung Countries"
px.histogram(x = beer_servings, y = countries, labels = {'x': 'Total Annual Beer Consumption, Liters', 'y': 'countries'},
             title = 'Histogram of Beer Consumption amoung Countries', nbins = 8)
```
